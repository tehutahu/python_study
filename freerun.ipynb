```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

np.random.seed(0)
```

```python
date_range = pd.date_range(start='2020-01-01', periods=365, freq='D')
date_range
```

```python
df = pd.DataFrame(data=np.random.randint(1, 31, 365), index=date_range, columns=['rand'])
df
```

```python
df['walking'] = np.random.randint(0, 100000, 365)
df
```

```python
df['eval'] = df['walking'].apply(lambda x:'S' if x > 50000 else 'A' if x > 5000 else 'B')
df
```

```python
fig, axes = plt.subplots(2,1, figsize=(16,8))
df['eval'].hist(ax=axes[0])
axes[1].hist(df['walking'].apply(lambda x:'50000' if x > 50000 else '5000' if x > 5000 else 'other'))
plt.show()

```
